```python
import matplotlib.pyplot as mpt
people=["kiran", "arun", "vijay", "varun"]

height=[145,151,165,173]

mpt.bar(people, height,color="green", width=0.5)

mpt.title("bar chart")

mpt.xlabel("age in years")

mpt.ylabel("height in meters")

mpt.show()
```
