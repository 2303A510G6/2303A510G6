4 - Data Preprocessing

Handling Categorical Features

```python
car_train = train_data.drop('binaryClass', axis=1)
car_labels = train_data[['binaryClass']]
```

```python
from sklearn.preprocessing import OrdinalEncoder

from sklearn.pipeline import Pipeline

pipe = Pipeline([('ord_enc', OrdinalEncoder())])

car_train_prepared = pipe.fit_transform(car_train)
```

```python
from sklearn.preprocessing import LabelEncoder

label_enc = LabelEncoder()

car_labels_prepared = label_enc.fit_transform(car_labels)
```

```python
car_train_prepared.shape, car_labels_prepared.shape
```

5 - Training Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

tree_clf = DecisionTreeClassifier()
tree_clf.set_params(class_weight='balanced', criterion='entropy')

tree_clf.fit(car_train_prepared, car_labels_prepared)
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(20,15))
plot_tree(tree_clf, feature_names=car_train.columns,
          class_names=['N','P']);
```

6 - Evaluating Decision Trees

```python
from sklearn.metrics import accuracy_score, classification_report,\
confusion_matrix

preds = tree_clf.predict(car_train_prepared)
print(f"Accuracy:{accuracy_score(car_labels_prepared, preds)}")
print(f"ClassReport:{classification_report(car_labels_prepared,preds)}")
print(f"Confusion Matrix:{confusion_matrix(car_labels_prepared,preds)}")
```

```python
# Let's hide warnings returned by grid search

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.model_selection import GridSearchCV

params_grid = {'max_leaf_nodes': list(range(0, 10)),
               'min_samples_split': [0,1,2, 3, 4],
               'min_samples_leaf': [0,1,2, 3, 4],
               'max_depth':[0,1,2,3,4,5],
               'max_features':[0,1,2,3,4,5],
              'max_leaf_nodes':[0,1,2,3,4,5]}

#refit is true by default. The best estimator is trained on the whole dataset

grid_search = GridSearchCV(DecisionTreeClassifier(random_state=42, class_weight='balanced'), params_grid, verbose=1, cv=3, refit=True)

grid_search.fit(car_train_prepared, car_labels_prepared)
```

```python
grid_search.best_params_
```

```python
grid_search.best_estimator_
```

```python
tree_best = grid_search.best_estimator_
```

```python
from sklearn.metrics import accuracy_score # Import the accuracy_score function

# Assuming your accuracy function was intended to calculate accuracy
# using the predicted and true labels:
def accuracy(X, model, y_true):
  """
  Calculates the accuracy of a model's predictions.

  Args:
    X: The input features.
    model: The trained model.
    y_true: The true labels.

  Returns:
    The accuracy score.
  """
  y_pred = model.predict(X)  # Get predictions from the model
  return accuracy_score(y_true, y_pred)  # Calculate and return accuracy

# Now you can call the accuracy function:
accuracy(car_train_prepared, tree_best, car_labels_prepared)
```

```python
from sklearn.metrics import confusion_matrix # Import confusion_matrix

# Assuming your conf_matrix function was intended to calculate and display
# the confusion matrix:
def conf_matrix(X, model, y_true):
  """
  Calculates and displays the confusion matrix.

  Args:
    X: The input features.
    model: The trained model.
    y_true: The true labels.
  """
  y_pred = model.predict(X)  # Get predictions from the model
  cm = confusion_matrix(y_true, y_pred)  # Calculate confusion matrix
  print(cm)  # Display the confusion matrix

# Now you can call the conf_matrix function:
conf_matrix(car_train_prepared, tree_best, car_labels_prepared)
```

```python
from sklearn.metrics import classification_report # Import classification_report

# Assuming your class_report function was intended to generate a classification report:
def class_report(X, model, y_true):
  """
  Generates and displays the classification report.

  Args:
    X: The input features.
    model: The trained model.
    y_true: The true labels.
  """
  y_pred = model.predict(X)  # Get predictions from the model
  report = classification_report(y_true, y_pred)  # Generate classification report
  print(report)  # Display the report

# Now you can call the class_report function:
class_report(car_train_prepared, tree_best, car_labels_prepared)
```
